```python
from pandas_profiling import ProfileReport
```

```python
pandasp_report = ProfileReport(apps, 
  dataset={ 
       'description': 'Almacena información de las aplicaciones móviles en Google Play Store',
       'author': 'Tu nombre'},
  variables={
       'descriptions': {
          'App': 'nombre de la aplicación', 
          'Category': 'categoría a la que pertenece',
          'Rating': 'calificación',
          'Reviews': 'revisiones',
          'Size': 'tamaño',
          'Installs': 'descargas',
          'Type': 'libre o de paga',
          'Price': 'precio',
          'Content Rating': 'segmento al que se dirige',
          'Genres': 'género',
          'Last Updated': 'fecha de última actualización',
          'Current Ver': 'versión actual',
          'Android Ver': 'versión del sistema operativo'}},
  interactions=None,
  correlations=None)

pandasp_report.to_notebook_iframe()
```
